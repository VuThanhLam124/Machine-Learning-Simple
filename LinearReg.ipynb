Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
# Tạo phương trình bậc 1
def f(x):
    return 2*x + 3
```

```python
x = np.linspace(-10, 10, 100)
y = f(x) + np.random.normal(1, 4, 100)
```

```python
plt.scatter(x, y)
```

```python
# dựa trên công thức của xác suất thống kê(môn MAS291 tại trường ĐH FPT), ta xây dựng hàm tìm w, b
def linear(x,y):
    w = 0
    b = 0
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    xy_mean = np.mean(x*y)
    x2_mean = np.mean(x**2)
    w = (x_mean*y_mean - xy_mean)/(x_mean**2 - x2_mean)
    b = y_mean - w*x_mean
    print("Best fit line: y = ", w, "x + ", b)
    print("loss: ", np.mean((w*x + b - y)**2))
    plt.plot(x, w*x + b, 'r')
    plt.scatter(x, y)
    plt.show()
```

```python
linear(x, y)
```

```python
# so sánh với thư viện sklearn
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x.reshape(-1, 1), y)
print("Best fit line: y = ", model.coef_[0], "x + ", model.intercept_)
print("loss: ", np.mean((model.coef_[0]*x + model.intercept_ - y)**2))
plt.plot(x, model.coef_[0]*x + model.intercept_, 'r')
plt.scatter(x, y)
plt.show()
```
